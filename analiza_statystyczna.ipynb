```python
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import f_oneway
from scipy.stats import shapiro
```

```python
file_path = "C:/Users/sarus/Downloads/statistics_data (5).xlsx"  # zamień na właściwą ścieżkę do pliku
```

```python
df = pd.read_excel(file_path, header = None)
df = df.dropna(axis=0, how='all')
df = df.drop(df.columns[0], axis=1)
# Przypisanie wierszy do kategorii
categories = df.iloc[:4].values.tolist()

# Przypisanie kategorii do nagłówków kolumn
df.columns = categories

# Usunięcie pierwszych czterech wierszy, które były wykorzystywane jako nagłówki kolumn
df = df.iloc[4:]

# Reset indeksów
df.reset_index(drop=True, inplace=True)
# Wyświetlenie wyników
#print(df)
```

```python
# Tokenizator: Default BERT tokenizer (uncased)
results_default_with_preprocessing = df['BBC']['With preprocessing']['Default BERT tokenizer (uncased)']['Hyperparameters default']
results_default_without_preprocessing = df['BBC']['Without preprocessing']['Default BERT tokenizer (uncased)']['Hyperparameters default']

# Test Shapiro-Wilka dla danych z i bez preprocessingu dla tokenizatora domyślnego
statistic, p_value_shapiro = shapiro(results_default_with_preprocessing)
print("Tokenizator: Default BERT tokenizer (uncased) - Z preprocessingu")
print("Wartość statystyki testu Shapiro-Wilka:", statistic)
print("Wartość p dla testu Shapiro-Wilka:", p_value_shapiro)
print()

statistic, p_value_shapiro = shapiro(results_default_without_preprocessing)
print("Tokenizator: Default BERT tokenizer (uncased) - Bez preprocessingu")
print("Wartość statystyki testu Shapiro-Wilka:", statistic)
print("Wartość p dla testu Shapiro-Wilka:", p_value_shapiro)
print()

# Tokenizator: Cased BERT tokenizer
results_cased_with_preprocessing = df['BBC']['With preprocessing']['Cased BERT tokenizer']['Hyperparameters default']
results_cased_without_preprocessing = df['BBC']['Without preprocessing']['Cased BERT tokenizer']['Hyperparameters default']

# Test Shapiro-Wilka dla danych z i bez preprocessingu dla cased BERT tokenizer
statistic, p_value_shapiro = shapiro(results_cased_with_preprocessing)
print("Tokenizator: Cased BERT tokenizer - Z preprocessingu")
print("Wartość statystyki testu Shapiro-Wilka:", statistic)
print("Wartość p dla testu Shapiro-Wilka:", p_value_shapiro)
print()

statistic, p_value_shapiro = shapiro(results_cased_without_preprocessing)
print("Tokenizator: Cased BERT tokenizer - Bez preprocessingu")
print("Wartość statystyki testu Shapiro-Wilka:", statistic)
print("Wartość p dla testu Shapiro-Wilka:", p_value_shapiro)
print()

# Tokenizator: Custom tokenizer
results_custom_with_preprocessing = df['BBC']['With preprocessing']['Custom Tokenizer']['Hyperparameters default']
results_custom_without_preprocessing = df['BBC']['Without preprocessing']['Custom Tokenizer']['Hyperparameters default']

# Test Shapiro-Wilka dla danych z i bez preprocessingu dla custom tokenizer
statistic, p_value_shapiro = shapiro(results_custom_with_preprocessing)
print("Tokenizator: Custom tokenizer - Z preprocessingu")
print("Wartość statystyki testu Shapiro-Wilka:", statistic)
print("Wartość p dla testu Shapiro-Wilka:", p_value_shapiro)
print()

statistic, p_value_shapiro = shapiro(results_custom_without_preprocessing)
print("Tokenizator: Custom tokenizer - Bez preprocessingu")
print("Wartość statystyki testu Shapiro-Wilka:", statistic)
print("Wartość p dla testu Shapiro-Wilka:", p_value_shapiro)
print()
```

```python

# Przypisanie nazw kolumn
categories = df.columns.levels[0]
preprocessing_options = df.columns.levels[1]
tokenizers = df.columns.levels[2]
hyperparameters = df.columns.levels[3]

# Przygotowanie pustej listy na wyniki ANOVA
anova_results = []

# Przeprowadzenie testu ANOVA dla wszystkich kombinacji
for tokenizer in tokenizers:
    for preprocessing in preprocessing_options:
        results = []
        for hyperparameter in hyperparameters:
            data = df['BBC'][preprocessing][tokenizer][hyperparameter]
            results.append(data)
        anova_result = f_oneway(*results)
        anova_results.append({
            'Tokenizer': tokenizer,
            'Preprocessing': preprocessing,
            'F-statistic': anova_result.statistic,
            'p-value': anova_result.pvalue
        })

# Tworzenie ramki danych z wynikami ANOVA
anova_df = pd.DataFrame(anova_results)

# Wyświetlenie wyników
print(anova_df)
```

```python
# Przypisanie nazw kolumn
categories_imdb = df.columns.levels[0]
preprocessing_options_imdb = df.columns.levels[1]
tokenizers_imdb = df.columns.levels[2]
hyperparameters_imdb = df.columns.levels[3]

# Przygotowanie pustej listy na wyniki ANOVA dla IMDb
anova_results_imdb = []

# Przeprowadzenie testu ANOVA dla wszystkich kombinacji dla IMDb
for tokenizer_imdb in tokenizers_imdb:
    for preprocessing_imdb in preprocessing_options_imdb:
        results_imdb = []
        for hyperparameter_imdb in hyperparameters_imdb:
            data_imdb = df['IMDb'][preprocessing_imdb][tokenizer_imdb][hyperparameter_imdb]
            results_imdb.append(data_imdb)
        anova_result_imdb = f_oneway(*results_imdb)
        anova_results_imdb.append({
            'Tokenizer': tokenizer_imdb,
            'Preprocessing': preprocessing_imdb,
            'F-statistic': anova_result_imdb.statistic,
            'p-value': anova_result_imdb.pvalue
        })

# Tworzenie ramki danych z wynikami ANOVA dla IMDb
anova_df_imdb = pd.DataFrame(anova_results_imdb)

# Wyświetlenie wyników dla IMDb
print(anova_df_imdb)
```

```python
plt.figure(figsize=(10, 6))
# Wykres słupkowy dla średniej wartości F-statystyki
sns.barplot(x='Tokenizer', y='F-statistic', hue='Preprocessing', data=anova_df)
plt.title('Średnia wartość F-statystyki dla Tokenizerów (BBC)')
plt.xlabel('Tokenizer')
plt.ylabel('Średnia wartość F-statystyki')
plt.legend(title='Preprocessing')
plt.show()

plt.figure(figsize=(10, 6))
# Wykres słupkowy dla średniej wartości F-statystyki dla IMDb
sns.barplot(x='Tokenizer', y='F-statistic', hue='Preprocessing', data=anova_df_imdb)
plt.title('Średnia wartość F-statystyki dla Tokenizerów (IMDb)')
plt.xlabel('Tokenizer')
plt.ylabel('Średnia wartość F-statystyki')
plt.legend(title='Preprocessing')
plt.show()
```

```python
# Wykres słupkowy dla wartości p
plt.figure(figsize=(10, 6))
sns.barplot(x='Tokenizer', y='p-value', hue='Preprocessing', data=anova_df)
plt.title('Wartość p dla Tokenizerów (BBC)')
plt.xlabel('Tokenizer')
plt.ylabel('Wartość p')
plt.legend(title='Preprocessing')
plt.show()
plt.figure(figsize=(10, 6))
# Wykres słupkowy dla wartości p dla IMDb
sns.barplot(x='Tokenizer', y='p-value', hue='Preprocessing', data=anova_df_imdb)
plt.title('Wartość p dla Tokenizerów (IMDb)')
plt.xlabel('Tokenizer')
plt.ylabel('Wartość p')
plt.legend(title='Preprocessing')
plt.show()
```
